```python
# 34 How to get all the dates corresponding to the month of July 2016?
start = date(2016, 7, 1)
end = date(2016, 8, 1)

dates = []
current = start

while (current < end):
	dates.append(current)
	current += timedelta(days=1)

print(dates)
```

```python
# 46  Create a structured array with x and y coordinates covering the [0,1]x[0,1] area 

x = np.linspace(0,1, 5)
y = np.linspace(0,1, 5)

x, y = np.meshgrid(x, y)

z = np.zeros(x.size, dtype=[('x', float), ('y', float)])

z['x'] = x.ravel()
z['y'] = y.ravel()

print(z)
```

```python
# 48

int_types = [np.int8, np.int16, np.int32, np.int64, np.uint, np.uint8, np.uint16, np.uint32, np.uint64]

print("Integer Types: ")
for t in int_types:
	info = np.iinfo(t)
	print(f"{t.__name__}: min={info.min}, max={info.max}")
	
```

```python
# for floating value

float_types = [np.float16, np.float32, np.float64]

print("Floating values: ")
for f in float_types:
	info = np.finfo(f)
	print(f"{f.__name__}: min={info.min}, max={info.max}")
```

```python
# 50  How to find the closest value (to a given scalar) in a vector?

vec = np.array([2.0, 4.2, 5.32, 6.87])
value = 5.11

closet = vec[np.abs(vec - value).argmin()]

print(closet)
```
